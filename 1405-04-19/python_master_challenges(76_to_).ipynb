## 76. با استفاده از Lambda لیستی از دیکشنری‌ها را بر اساس چند کلید مرتب کنید.

```python
list_of_dicts = [
    {'name': '10', 'id': 3},
    {'name': '15', 'id': 0},
    {'name': '05', 'id': 4},
    {'name': '35', 'id': 2},
    {'name': '09', 'id': 1},
]

print(list_of_dicts)
list_of_dicts.sort(
    key=lambda e: (e['name'], e['id'],)
)
print(list_of_dicts)
```
