## 82. پرتکرارترین کلمه یک فایل را پیدا کنید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).parent

class ImprovedFileAnalyzer:
    file_info = {
    "line_count": 0,
    "words_count": 0,
    "char_count": 0,
    "words": {},
    }
    _cache = {}

    def __init__(self) -> None:
        """
        A class to work with txt files and analyze them.
        """
        pass

    def analyze(self, target_fp: Path) -> dict[str, int]:
        """
        Analyze the file and return info.

        Args:
            target_fp (Path): Path object to target file for analyze.

        Returns:
            dict[str, int]
        """
        with open(target_fp, 'r', encoding='utf-8') as f:
            while line := f.readline():
                self.file_info["line_count"] += 1
                self.file_info["words_count"] += len(line.split())
                self.file_info["char_count"] += len(line)

                words = line.split()
                for w in words:
                    if w not in self.file_info['words']:
                        self.file_info['words'][f"{w}"] = 1
                    else: 
                        self.file_info['words'][f"{w}"] += 1
        
        self._cache[target_fp.name] = self.file_info
        return self.file_info

    def get_max_repeated(self, data: dict) -> dict[str, int]:
        """
        Get most repeated word from a data dict of words.

        Args:
            data (dict): which returned from analyze() and cached in _cache.

        Returns:
            dict[str, int]
        """
        max_repeat = {'repeat': 0}
        for k, v in data['words'].items():
            if v > max_repeat['repeat']:
                max_repeat = dict(word=k, repeat=v)

        return max_repeat

    def __str__(self):
        """
        Override __str__() dunder to prepare a nice report of the files which this object has been analyzed and cached.
        """
        info = []
        for k, v in self._cache.items():
            string = f"\n---file info---\n* file name => '{k}'\n---\n* how many lines? => '{v["line_count"]}'\n---\n* how many words? => '{v["words_count"]}'\n---\n* how many characters? => '{v["char_count"]}'\n---\n* word with max use => \n\t{self.get_max_repeated(v)}\n---------------\n\n"
            info.append(string)
        return "".join(info)


target_fp = BASE_DIR / 'test_file_exercise_81.txt'
file_analyzer = ImprovedFileAnalyzer()
file_analyzer.analyze(target_fp)

print(file_analyzer)
# OR
print(
    file_analyzer.get_max_repeated(
        file_analyzer._cache[f'{target_fp.name}']
    )
)
```

## 83. تمام خطوط تکراری فایل را حذف کنید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).parent

class ImprovedFileAnalyzer:
    file_info = {
    "line_count": 0,
    "words_count": 0,
    "char_count": 0,
    "words": {},
    }
    _cache = {}

    def __init__(self) -> None:
        """
        A class to work with txt files and analyze them.
        """
        pass

    def analyze(self, target_fp: Path) -> dict[str, int]:
        """
        Analyze the file and return info.

        Args:
            target_fp (Path): Path object to target file for analyze.

        Returns:
            dict[str, int]
        """
        with open(target_fp, 'r', encoding='utf-8') as f:
            while line := f.readline():
                self.file_info["line_count"] += 1
                self.file_info["words_count"] += len(line.split())
                self.file_info["char_count"] += len(line)

                words = line.split()
                for w in words:
                    if w not in self.file_info['words']:
                        self.file_info['words'][f"{w}"] = 1
                    else: 
                        self.file_info['words'][f"{w}"] += 1
        
        self._cache[target_fp.name] = self.file_info
        return self.file_info

    def get_max_repeated(self, data: dict) -> dict[str, int]:
        """
        Get most repeated word from a data dict of words.

        Args:
            data (dict): which returned from analyze() and cached in _cache.

        Returns:
            dict[str, int]
        """
        max_repeat = {'repeat': 0}
        for k, v in data['words'].items():
            if v > max_repeat['repeat']:
                max_repeat = dict(word=k, repeat=v)

        return max_repeat

    def remove_repeated_line(self, target_fp: Path) -> None:
        """
        Detect repeated lines in a file and make a clean version of file without them. use set to detect repeated lines. if len(set) didn't increased, we have a repeated line!

        Args:
            target_fp (Path): Path object to target file.

        Returns:
            None
        """
        clean_fp = target_fp.parent / f"clean_{target_fp.name}"
        with open(target_fp, 'r', encoding='utf-8') as f, open(clean_fp, 'w', encoding='utf-8') as t:
            check_set = set()
            while line := f.readline():
                before_len = len(check_set)
                check_set.add(line)
                if not len(check_set) == before_len:
                    t.writelines([line])
                else: 
                    continue
        
        print(f"{target_fp} cleaned in {clean_fp} ✅.")

    def __str__(self):
        """
        Override __str__() dunder to prepare a nice report of the files which this object has been analyzed and cached.
        """
        info = []
        for k, v in self._cache.items():
            string = f"\n---file info---\n* file name => '{k}'\n---\n* how many lines? => '{v["line_count"]}'\n---\n* how many words? => '{v["words_count"]}'\n---\n* how many characters? => '{v["char_count"]}'\n---\n* word with max use => \n\t{self.get_max_repeated(v)}\n---------------\n\n"
            info.append(string)
        return "".join(info)


target_fp = BASE_DIR / 'test_file_exercise_83.txt'
file_analyzer = ImprovedFileAnalyzer()
file_analyzer.analyze(target_fp)

file_analyzer.remove_repeated_line(target_fp)
```

## 84. دو فایل متنی را با هم مقایسه کنید و خطوط متفاوت را نمایش دهید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).parent


def differences_of_two_file(file1: Path, file2: Path):
    """
    Get two filepath object, open and print different lines.
    """
    with (
        open(file1, 'r') as f1, 
        open(file2, 'r') as f2
    ):
        line_count = 0
        while (l1 := f1.readline(), l2 := f2.readline()):
            if l1 == '':
                print(f"📁📕'{f1.name}' file end! ")
            if l2 == '':
                print(f"📁📕'{f2.name}' file end! ")
            
            if l1 == '' and l2 == '':
                break

            check_set = set()
            check_set.add(l1)
            check_set.add(l2)
            if len(check_set) > 1:
                print(f"line number:'{line_count}'\n⛔ {f1.name}:\n\t>> {l1}\n⛔ {f2.name}:\n\t>> {l2}\n")
            elif len(check_set) == 1:
                print(f"line number: '{line_count}'\n SAME ✅\n")
            line_count += 1


f1_fp = BASE_DIR / 'test_file_exercise_84_1.txt'
f2_fp = BASE_DIR / 'test_file_exercise_84_2.txt'
differences_of_two_file(f1_fp, f2_fp)
```
