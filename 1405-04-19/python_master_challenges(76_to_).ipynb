## 82. پرتکرارترین کلمه یک فایل را پیدا کنید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).parent

class ImprovedFileAnalyzer:
    file_info = {
    "line_count": 0,
    "words_count": 0,
    "char_count": 0,
    "words": {},
    }
    _cache = {}

    def __init__(self) -> None:
        """
        A class to work with txt files and analyze them.
        """
        pass

    def analyze(self, target_fp: Path) -> dict[str, int]:
        """
        Analyze the file and return info.

        Args:
            target_fp (Path): Path object to target file for analyze.

        Returns:
            dict[str, int]
        """
        with open(target_fp, 'r', encoding='utf-8') as f:
            while line := f.readline():
                self.file_info["line_count"] += 1
                self.file_info["words_count"] += len(line.split())
                self.file_info["char_count"] += len(line)

                words = line.split()
                for w in words:
                    if w not in self.file_info['words']:
                        self.file_info['words'][f"{w}"] = 1
                    else: 
                        self.file_info['words'][f"{w}"] += 1
        
        self._cache[target_fp.name] = self.file_info
        return self.file_info

    def get_max_repeated(self, data: dict) -> dict[str, int]:
        """
        Get most repeated word from a data dict of words.

        Args:
            data (dict): which returned from analyze() and cached in _cache.

        Returns:
            dict[str, int]
        """
        max_repeat = {'repeat': 0}
        for k, v in data['words'].items():
            if v > max_repeat['repeat']:
                max_repeat = dict(word=k, repeat=v)

        return max_repeat

    def __str__(self):
        """
        Override __str__() dunder to prepare a nice report of the files which this object has been analyzed and cached.
        """
        info = []
        for k, v in self._cache.items():
            string = f"\n---file info---\n* file name => '{k}'\n---\n* how many lines? => '{v["line_count"]}'\n---\n* how many words? => '{v["words_count"]}'\n---\n* how many characters? => '{v["char_count"]}'\n---\n* word with max use => \n\t{self.get_max_repeated(v)}\n---------------\n\n"
            info.append(string)
        return "".join(info)


target_fp = BASE_DIR / 'test_file_exercise_81.txt'
file_analyzer = ImprovedFileAnalyzer()
file_analyzer.analyze(target_fp)

print(file_analyzer)
# OR
print(
    file_analyzer.get_max_repeated(
        file_analyzer._cache[f'{target_fp.name}']
    )
)
```

## 83. تمام خطوط تکراری فایل را حذف کنید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).parent

class ImprovedFileAnalyzer:
    file_info = {
    "line_count": 0,
    "words_count": 0,
    "char_count": 0,
    "words": {},
    }
    _cache = {}

    def __init__(self) -> None:
        """
        A class to work with txt files and analyze them.
        """
        pass

    def analyze(self, target_fp: Path) -> dict[str, int]:
        """
        Analyze the file and return info.

        Args:
            target_fp (Path): Path object to target file for analyze.

        Returns:
            dict[str, int]
        """
        with open(target_fp, 'r', encoding='utf-8') as f:
            while line := f.readline():
                self.file_info["line_count"] += 1
                self.file_info["words_count"] += len(line.split())
                self.file_info["char_count"] += len(line)

                words = line.split()
                for w in words:
                    if w not in self.file_info['words']:
                        self.file_info['words'][f"{w}"] = 1
                    else: 
                        self.file_info['words'][f"{w}"] += 1
        
        self._cache[target_fp.name] = self.file_info
        return self.file_info

    def get_max_repeated(self, data: dict) -> dict[str, int]:
        """
        Get most repeated word from a data dict of words.

        Args:
            data (dict): which returned from analyze() and cached in _cache.

        Returns:
            dict[str, int]
        """
        max_repeat = {'repeat': 0}
        for k, v in data['words'].items():
            if v > max_repeat['repeat']:
                max_repeat = dict(word=k, repeat=v)

        return max_repeat

    def remove_repeated_line(self, target_fp: Path) -> None:
        """
        Detect repeated lines in a file and make a clean version of file without them. use set to detect repeated lines. if len(set) didn't increased, we have a repeated line!

        Args:
            target_fp (Path): Path object to target file.

        Returns:
            None
        """
        clean_fp = target_fp.parent / f"clean_{target_fp.name}"
        with open(target_fp, 'r', encoding='utf-8') as f, open(clean_fp, 'w', encoding='utf-8') as t:
            check_set = set()
            while line := f.readline():
                before_len = len(check_set)
                check_set.add(line)
                if not len(check_set) == before_len:
                    t.writelines([line])
                else: 
                    continue
        
        print(f"{target_fp} cleaned in {clean_fp} ✅.")

    def __str__(self):
        """
        Override __str__() dunder to prepare a nice report of the files which this object has been analyzed and cached.
        """
        info = []
        for k, v in self._cache.items():
            string = f"\n---file info---\n* file name => '{k}'\n---\n* how many lines? => '{v["line_count"]}'\n---\n* how many words? => '{v["words_count"]}'\n---\n* how many characters? => '{v["char_count"]}'\n---\n* word with max use => \n\t{self.get_max_repeated(v)}\n---------------\n\n"
            info.append(string)
        return "".join(info)


target_fp = BASE_DIR / 'test_file_exercise_83.txt'
file_analyzer = ImprovedFileAnalyzer()
file_analyzer.analyze(target_fp)

file_analyzer.remove_repeated_line(target_fp)
```

## 84. دو فایل متنی را با هم مقایسه کنید و خطوط متفاوت را نمایش دهید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).parent


def differences_of_two_file(file1: Path, file2: Path):
    """
    Get two filepath object, open and print different lines.
    """
    with (
        open(file1, 'r') as f1, 
        open(file2, 'r') as f2
    ):
        line_count = 0
        while (l1 := f1.readline(), l2 := f2.readline()):
            if l1 == '':
                print(f"📁📕'{f1.name}' file end! ")
            if l2 == '':
                print(f"📁📕'{f2.name}' file end! ")
            
            if l1 == '' and l2 == '':
                break

            check_set = set()
            check_set.add(l1)
            check_set.add(l2)
            if len(check_set) > 1:
                print(f"line number:'{line_count}'\n⛔ {f1.name}:\n\t>> {l1}\n⛔ {f2.name}:\n\t>> {l2}\n")
            elif len(check_set) == 1:
                print(f"line number: '{line_count}'\n SAME ✅\n")
            line_count += 1


f1_fp = BASE_DIR / 'test_file_exercise_84_1.txt'
f2_fp = BASE_DIR / 'test_file_exercise_84_2.txt'
differences_of_two_file(f1_fp, f2_fp)
```

## 88. پوشه‌ای را به صورت بازگشتی کپی کنید.

```python
from pathlib import Path
import shutil
BASE_DIR = Path(__name__).parent.resolve()

def recursive_copy_dir(src_dir: Path, dest_dir: Path) -> Path:
    """
    Copy a directory with all files in it (without any ignore patterns for now) into a destination path. using `shutil.copytree()`_.

    Args:
        src_dir (Path): Path object which must be a directory to perform copytree().
        dest_dir (Path): Path object which must be a directory to copy into it.

    Returns:
        Path: if performance was successful return directory path object of copied directory.

    .. _shutil.copytree():
        https://docs.python.org/3/library/shutil.html#shutil.copytree
    """
    dest_dir = dest_dir / src_dir.name
    dest_dir.mkdir(exist_ok=True)
    if not src_dir.exists() or not src_dir.is_dir():
        raise AttributeError(f"{src_dir} does not exist or not a directory.\nError message")
    else:
        shutil.copytree(
            src=src_dir,
            dst=dest_dir,
            ignore=None,
            dirs_exist_ok=True,
        )
        return Path(dest_dir)


# Try to copy first directory of this repo to current directory.
src_directory = BASE_DIR.parent / '1405-04-06'
dest_directory = BASE_DIR

result = recursive_copy_dir(
    src_directory,
    dest_directory,
)
print(result)
```

## 89. حجم تمام فایل‌های یک پوشه را محاسبه کنید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).resolve().parent

def get_dir_size(target_dir: Path) -> dict[str, float | dict[str, float]]:
    """
    Calcluate each file size and a overall size of a directory. using `Path().stats().st_size`_

    Args:
        target_dir (Path): Path object to where we want size details.

    Returns:
        dict[str, float | dict[str, float]]: include each file size and overall size of directory. unit is KB.

    Examples:
    >>> get_dir_size(BASE_DIR)
    {
        'overall_KB': 35.62,
        'files_KB': {
            'python_master_challenges(76_to_).ipynb': 25.06,
            'README.md': 3.0,
            'test_file_exercise_81.txt': 7.12,
            'test_file_exercise_83.txt': 0.3,
            'test_file_exercise_84_1.txt': 0.07,
            'test_file_exercise_84_2.txt': 0.07
        }
    }

    .. _Path().stats().st_size:
        https://docs.python.org/3/library/pathlib.html#pathlib.Path.stat
    """

    r_dict = {}
    r_dict['overall_KB'] = 0 
    r_dict['files_KB'] = {}
    for f in target_dir.rglob('*'):
        r_dict['files_KB'][f.name] = round(f.stat().st_size / 1024, 2)
        r_dict['overall_KB'] += r_dict['files_KB'][f.name]
    return r_dict


get_dir_size(BASE_DIR)
```

## 90. تمام فایل‌های خالی یک پوشه را پیدا کنید.

```python
from pathlib import Path
BASE_DIR = Path(__name__).resolve().parent


def detect_empty_files(target_dir: Path) -> list[Path]:
    """
    Detect and return a list of filepath objects for empty files with exactly zero byte size.

    Args:
        target_dir (Path): Path object to where we want check for.

    Returns:
        list[Path]
    """
    empty_files = []
    for f in target_dir.rglob('*'):
        if f.stat().st_size == 0:
            empty_files.append(f)
    
    return empty_files


target_dir = BASE_DIR / 'test_dir_exercise_90'
detect_empty_files(target_dir)
```

## 91. یک فایل JSON را بخوانید و ساختار آن را اعتبارسنجی کنید.

```python
import requests
import json
from pathlib import Path
BASE_DIR = Path(__name__).resolve().parent


def get_cor() -> str:
    """
    Get current coordinate of ISS from api.

    Note:
        Copied from my repo (`elastipy/Day033`_)

    Returns:
        str

    .. _elastipy/Day033:
        https://github.com/erfaw/elasti_py/blob/master/Day033-erf/ISS_tracker/iss_track.py
    """
    try:
        res_iss = requests.get("http://api.open-notify.org/iss-now.json")
        res_iss.raise_for_status()

    except requests.exceptions.ConnectTimeout as er_ms:
        print(f'Error: request timed out!\nerror message:\n{er_ms}')
    return res_iss.text


def is_json(str: str) -> bool:
    """
    Accept a string sequence and validate it.

    Args:
        str (str): which we want to check for.

    Returns:
        bool
    """
    try:
        json.loads(str)
        return True
    except json.JSONDecodeError:
        return False 


json_result = get_cor()
print(is_json(json_result))

with open(BASE_DIR/'test_file_exercise_84_2.txt', 'r') as f:
    print(
        is_json(f.read())
    )
```
