## 46. تمام اعداد Armstrong بین 1 تا 100000 را پیدا کنید.

### What is `Armstrong Number` ?
> ChatGPT:
>> Armstrong numbers (also called Narcissistic numbers) are numbers that are <u>equal to the sum of their own digits</u>, where <u>each digit is raised to the power of the number of digits</u>.    
>> Examples:    
>> 153    
>> = 1³ + 5³ + 3³    
>> = 1 + 125 + 27    
>> = 153    


````python
def improved_manual_split(seq: str, by: str | None=" ") -> list[str]: 
    """
    Try to implement functionality of buil-in `str.split()` in python. 

    Note:
        in addition of original function, it could split by None, which means split by each char.

    Args: 
        seq (str): a string sequence.
        by (str): a character to split by. (default to ' ' empty space.)

    Returns: 
        list[str]: list of splited by strings.
    """
    splited_without_split_func = []
    word=""
    for c in seq:
        if by is None:
            splited_without_split_func.append(c)
            continue

        if not c == by:
            word += c
        else: 
            splited_without_split_func.append(word)
            word=""
    splited_without_split_func.append(word)

    if splited_without_split_func[-1] == '':
        splited_without_split_func.pop()

    return splited_without_split_func


def is_armstrong_num(num: int) -> bool:
    """
    Check if a number is armstrong or not.

    Args:
        num (int):

    Returns:
        bool
    """
    num_str = str(num)
    splited = improved_manual_split(num_str, None)

    sum = 0
    for n in splited:
        number = int(n)
        sum += number**(len(splited))
    
    if num == sum:
        return True
    else:
        return False


armstrongs_in_range = []
num_range = range(1, 100000+1)
for n_ in num_range:
    if is_armstrong_num(n_):
        armstrongs_in_range.append(n_)
    else: 
        continue

print(
    f"Armstrong numbers in range 1 to 100,000:\n  >> {armstrongs_in_range}"
)
````

## 47. تمام اعداد Perfect بین 1 تا 10000 را پیدا کنید.

### What is `Perfect Number` ?
> Google Search
>> In number theory, a perfect number is a <u>positive integer that is equal to the sum of its positive proper divisors</u>, that is, divisors excluding the number itself.    
>> It has to return True with this example values [6, 28, 496, 8128]    

````python
def is_perfect_number(n) -> bool:
    """
    checks a number is perfect or not.

    Note:
        Copied from my solution in `1405-04-06/01-check-perfect-num.py`_

    Args:
        n (int):

    Returns:
        bool

    .. _1405-04-06/01-check-perfect-num.py:
        https://github.com/erfaw/aa-internship-homeworks/blob/main/1405-04-06/01-check-perfect-num.py
    """
    # Check if input is a numeric value
    try: 
        number = float(n)
    except ValueError: 
        print("Please insert a valid input. (integer number)")

    # Check inputted number to be int
    if not number == int(number):
        return False
    else: 
        number = int(number)

    # Check for positivity
    if not number > 0:
        return False
    
    # End check to be perfect 
    sum = 0
    for i in range(1, number): 
        if number % i == 0:
            sum += i
    
    if not number == sum :
        return False

    return True


perfects_in_range = []
num_range = range(1, 10000+1)
for n_ in num_range:
    if is_perfect_number(n_):
        perfects_in_range.append(n_)

print(
    f"Perfect numbers in range 1 to 10,000:\n  >> {perfects_in_range}"
)
````

## 48. تمام اعداد Strong Number را پیدا کنید.

### What is `Strong Number` ?
> ChatGPT
>> A **Strong number** is a number whose value equals the **sum of the factorials of its digits**.    
>>     
>> Example:    
>>        
>> 145    
>> = 1! + 4! + 5!    
>> = 1 + 24 + 120    
>> = 145    
>> Common Strong numbers in base 10 are:       
>> ```text id="9vk22j"    
>> 1, 2, 145, 40585    
>> ```    


````python
def calculate_factorial(n: int) -> int:
    """
    Calculate factorial of recieved number. with recursive way.

    Note:
        Copied from my solution in `1405-04-06/02-sum-factorial.py`_

    Args:
        n (int):

    Returns:
        int

    .. _1405-04-06/02-sum-factorial.py:
        https://github.com/erfaw/aa-internship-homeworks/blob/main/1405-04-06/02-sum-factorial.py
    """
    if n >= 1:
        return n * calculate_factorial(n - 1)
    else:
        return 1


def improved_manual_split(seq: str, by: str | None = " ") -> list[str]:
    """
    Try to implement functionality of buil-in `str.split()` in python.

    Note:
        in addition of original function, it could split by None, which means split by each char.

    Args:
        seq (str): a string sequence.
        by (str): a character to split by. (default to ' ' empty space.)

    Returns:
        list[str]: list of splited by strings.
    """
    splited_without_split_func = []
    word = ""
    for c in seq:
        if by is None:
            splited_without_split_func.append(c)
            continue

        if not c == by:
            word += c
        else:
            splited_without_split_func.append(word)
            word = ""
    splited_without_split_func.append(word)

    if splited_without_split_func[-1] == "":
        splited_without_split_func.pop()

    return splited_without_split_func


def is_strong_num(num: int) -> bool:
    """
    Check if a number is strong or not.

    Args:
        num (int):

    Returns:
        bool
    """
    num_str = str(num)
    splited = improved_manual_split(num_str, None)

    sum = 0
    for n in splited:
        number = int(n)
        sum += calculate_factorial(number)

    if num == sum:
        return True
    else:
        return False


strongs_in_range = []
num_range = range(1, 50000+1)
for n_ in num_range:
    if is_strong_num(n_):
        strongs_in_range.append(n_)

print(
    f"Strong numbers in range 1 to 1000:\n  >> {strongs_in_range}"
)
````

## 49. بررسی کنید یک عدد Happy Number است یا خیر.

### What is `Happy Number` ?
> ChatGPT
>> A **Happy number** is a number that eventually reaches **1** when you repeatedly replace it with the **sum of the squares of its digits**.    
>>     
>> Example (`19`):    
>>     
>> ```text id="6qoqvn"    
>> 19 → 1²+9² = 82    
>> 82 → 8²+2² = 68    
>> 68 → 6²+8² = 100    
>> 100 → 1²+0²+0² = 1    
>> ```    
>>     
>> Since it reaches **1**, `19` is a **happy number**.    
>>     
>> If the process falls into a repeating cycle that **never reaches 1**, the number is called an **unhappy (or sad) number**.    
>> 

````python
def is_happy_num(num: int, tries: int = 3) -> bool:
    """
    Check if a number is happy or not.

    Args:
        num (int):
        tries (int): defines how many cycle to check. (if not send as arg use 3).

    Returns:
        bool
    """

    def cycle(n_):
        num_str = str(n_)
        splited = improved_manual_split(num_str, None)

        sum = 0
        for n in splited:
            number = int(n)
            sum += number**2
        return sum

    sum_ = num
    for _ in range(tries + 1):
        r = cycle(sum_)
        if r == 1:
            return True
        else:
            sum_ = r

    return False


happies_in_range = []
num_range = range(1, 1000+1)
for n in num_range:
    if is_happy_num(n,1):
        happies_in_range.append(n)


print(
    f"Happy numbers in range 1 to 1000 (with default 3 tries):\n  >> {happies_in_range}\n  len of: {len(happies_in_range)}"
)
# Its intresting to know there is a semi-pattern in happy numbers with 1 try somehow :) 01 -> 10, 13 -> 31, 68 -> 86, ...
````

## 50. مسئله کلاسیک FizzBuzz را پیاده‌سازی کنید، اما به‌گونه‌ای که تعداد قوانین قابل تغییر و توسعه باشد (قابل افزودن قوانین جدید بدون تغییر منطق اصلی برنامه).

### What is `FizzBuzz` ? 
> ChatGPT
>> The task is:    
>> * Print numbers from **1 to N**.    
>> * If a number is divisible by **3**, print **"Fizz"**.    
>> * If it's divisible by **5**, print **"Buzz"**.    
>> * If it's divisible by **both 3 and 5**, print **"FizzBuzz"**.    
>> * Otherwise, print the number itself.    
>>     
>> Example for `1..15`:     
>> 1 2 Fizz 4 Buzz Fizz 7 8 Fizz Buzz 11 Fizz 13 14 FizzBuzz    

````python
def run_fizz_buzz(num: int) -> None:
    """
    Run a while loop to num with FizzBuzz.

    Args:
        num (int): number to stop the loop.

    Returns:
        None
    """
    i = 1
    while i <= num:
        if i%3==0 and i%5==0:
            print('FizzBuzz', end=' ,')
        elif i % 3 == 0:
            print('Fizz', end=' ,')
        elif i % 5 == 0:
            print('Buzz', end=' ,')
        else:
            print(i, end=' ,')
        i+=1
    print('Poof...')


run_fizz_buzz(20)
````
