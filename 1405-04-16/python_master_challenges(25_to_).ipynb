## 25. لیست را با Insertion Sort مرتب کنید.

### `Insertion Sort`
> ChatGPT:    
>> **Insertion Sort** is a comparison-based sorting algorithm that builds the sorted portion of a list one element at a time. It repeatedly takes the next element from the unsorted part and inserts it into its correct position within the sorted part by shifting larger elements to the right. It is simple to implement, works in-place, and performs well on small or nearly sorted datasets.


```python
input_list = [5, 2, 4, 6, 1, 3]

def insertion_sort(l: list) -> list:
    """
    Trying to implement Insertion Sort manually.

    Args:
        l (list): unsorted list to sort.

    Returns:
        list
    """
    ordered_part = []
    unordered_part = l.copy()

    if len(ordered_part) == 0:
        ordered_part.append(unordered_part[0])
    
    for number in unordered_part:
        if len(ordered_part) == 1:
            if ordered_part[-1] < number:
                ordered_part.append(number)
            elif number < ordered_part[-1]:
                ordered_part.insert(0, number)
        else: 
            if number < ordered_part[0]:
                ordered_part.insert(0, number)
            elif ordered_part[-1] < number:
                ordered_part.append(number)
            else: 
                for n in ordered_part:
                    if n > number:
                        index = ordered_part.index(n)
                        ordered_part.insert(index, number)
                        break

    return ordered_part


insertion_sort(input_list)
```

```python
def is_sub(list: list, sub: list) -> bool:
    """
    Get two list and check if sub is subset of list or not. 

    Args:
        list (list):
        sub (list):

    Returns:
        bool
    """
    if len(sub) == 1:
        if sub[0] in list:
            return True
        else:
            return False
    else:
        for i in range(len(sub)):
            if i == 0:
                try:
                    first_index_in_list = list.index(sub[i])
                except ValueError:
                    return False

            else: # i = 1
                if list[first_index_in_list+i] == sub[i]:
                    continue
                else: 
                    return False
        return True


list1 = [1,2,3,4]
list2 = [2,3]
is_sub(list1, list2)
```

```python
def manual_reverse(l: list) -> list:
    """
    Get a list and reverse its order.

    Args:
        l (list): 

    Returns:
        list
    """
    r_list = l.copy()
    reversed_range = range(len(l)-1, -1, -1)
    for i in range(len(l)):
        r_list[i] = l[reversed_range[i]]

    return r_list


input_list = [1,2,3,4,5,6,7]
print(f"original list: {input_list}\n---\nreversed list: {manual_reverse(input_list)}")
```
