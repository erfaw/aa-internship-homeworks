## 63. تابع بازگشتی برای برعکس کردن رشته بنویسید.

```python
def r_string(str):
    """
    Reverse a string sequence with recursion.

    Args:
        s (str):
    
    Returns:
        str
    """
    if len(str) == 0:
        return ''
    return r_string(str[1:]) + str[0]


r_string('hello')
```

## 65. تابع بازگشتی برای جستجو در یک لیست بنویسید.

```python
def recursive_search(l: list, s: int) -> bool:
    """
    Search for existance of an element in a list.

    Args:
        l (list): to search into.
        s (int): element we are looking for.

    Returns:
        bool
    """
    if l[0] == s:
        return True
    else:
        try: 
            return recursive_search(l[1:], s)
        except IndexError:
            return False


recursive_search([1,2,3,4,5], 6)
```

## 66. یک Decorator بنویسید که زمان اجرای هر تابع را اندازه‌گیری کند.

```python
from collections.abc import Callable
import time

def runtime(func: Callable):
    """
    Decorator to print runtime in seconds after run end.
    """
    def wrapper(n):
        start = time.time()
        func(n)
        end = time.time()
        print(f"Runtime: {end-start}")
    
    return wrapper


@runtime
def a_test_func(n):
    for i in range(n+1):
        print(i)
        time.sleep(n/2)


a_test_func(3)
```

## 67. یک Decorator بنویسید که تعداد دفعات فراخوانی یک تابع را ثبت کند.

```python
from collections.abc import Callable
_count = 0
def call_count(func: Callable):
    def wrapper():
        global _count
        func()
        _count += 1
    return wrapper

@call_count
def a_test_func():
    print('nothing...')

print(f"count is: {_count}")

for _ in range(5):
    a_test_func()

print(f"count is: {_count}")
```

## 68. یک Decorator بنویسید که خروجی تابع را Cache کند.

```python
_cache = dict()


def cache_result(func):
    """
    Decorator to cache result of each function run in _cache dict.
    """

    def wrapper(i):
        func_name = getattr(func, "__name__")
        result = func(i)
        if func_name not in _cache:
            _cache[func_name] = {}
        _cache[func_name][i] = result
        return result

    return wrapper


@cache_result
def a_test_func(i):
    return i**2

for i in range(5+1):
    a_test_func(i)

_cache
```

## 69. یک Decorator بنویسید که فقط به کاربران Admin اجازه اجرای تابع را بدهد.

```python
def admin_only(func):
    def wrapper(user):
        if user['is_admin']:
            return func(user)
        else:
            return "Need admin privileges"
    return wrapper

user1 = {'name': 'Erfan',}
user1['is_admin'] = True # type: ignore

@admin_only
def restart_whole_service(user):
    print(f'restarting by {user['name']}')

restart_whole_service(user1)
```

## 70. یک Decorator بنویسید که تمام آرگومان‌های ورودی را Log کند.

```python
from logging import getLogger, basicConfig, INFO
from pathlib import Path
BASE_DIR = Path(__name__).parent
logger = getLogger(__name__)
basicConfig(
    filename=(BASE_DIR / 'exercise_70_logs.txt'),
    filemode='a',
    level=INFO,
    encoding='utf-8',
    datefmt=r"%Y-%m-%dT%I:%M:%S%p",
    format=r"%(asctime)s:%(name)s:%(message)s",
)


def log_args(func):
    def wrapper(i):
        func_name = getattr(func, "__name__")
        func(i)
        message = f"{func_name} called with '{i}' arg."
        logger.info(msg=message)
    return wrapper


@log_args
def a_test_func(i):
    return i**2

for i in range(5+1):
    a_test_func(i)
```

## 71. یک Generator برای تولید اعداد اول بنویسید.

```python
def generator_prime_num():
    """
    Returns a Generator(iterator) for prime numbers.
    """
    i = 1
    while True:
        if i == 1:
            yield 1
        if not i <= 1:
            for n in range(2,i):
                if i % n == 0:
                    break
            else:
                yield i
        i += 1

prime_numbers = generator_prime_num()

n = 100
for _ in range(n+1):
    print(
        next(prime_numbers), end=',  '
    )
```

## 72. یک Generator برای تولید اعداد فیبوناچی بنویسید.

```python
def generator_fibonacci_sequence():
    """
    Returns a Generator(iterator) for Fibonacci sequence.
    """
    prev_num = 0
    current_num = 1

    while True:
        result = current_num + prev_num
        yield result
        prev_num, current_num = current_num, result 


fibonacci_seq = generator_fibonacci_sequence()

n = 10
for _ in range(n+1):
    print(
        next(fibonacci_seq), end=',  '
    )
```

## 73. یک Generator برای خواندن فایل‌های بزرگ به صورت خط‌به‌خط بنویسید.

```python
from pathlib import Path

BASE_DIR = Path(__name__).parent


def generator_readline(filepath: Path):
    """
    Returns a Generator(iterator) to read big files line by line.

    Args:
        filepath (Path):
    """
    with open(filepath, "r") as file:

        while True:

            result = file.readline()
            yield result

            if result == "":
                yield "---\n---\n**File has been read completely**"
                break


test_file_fp = BASE_DIR / "test_file_exercise_73.txt"
readline_gen = generator_readline(test_file_fp)

# n = 500
# for _ in range(n+1):
#     print(
#         next(readline_gen)
#     )

for line in readline_gen:
    print(line)
```

## 75. یک Generator بنویسید که اعداد زوج نامتناهی تولید کند.

```python
def generator_even_num():
    num = 0
    while True:
        yield num
        num += 2

even_num_gen = generator_even_num()

n = 20
for i in range(n+1):
    print(
        next(even_num_gen),
        end=",  "
    )
```
