### What is `Permutation`
> ChatGPT    
>> **Permutation** is an arrangement of elements where **the order matters**.    
>> For the letters `ABC`, the permutations are:    
>> `ABC`, `ACB`, `BAC`, `BCA`, `CAB`, `CBA`.    
>> If two arrangements differ only in order (e.g., `ABC` vs. `BAC`), they are considered different permutations.    
>> | Situation                      | Formula                     |
>> | ------------------------------ | --------------------------- |
>> | All distinct, use all elements | (n!)                        |
>> | Repeated elements              | (\frac{n!}{n_1!n_2!\cdots}) |
>> | Use only (k) of (n) elements   | (\frac{n!}{(n-k)!})         |

```python
# My Failure tries

string_1 = "abc"
permutation_set = set()

str_list = list(string_1)

for char in str_list:

    tmp_list = str_list.copy()

    for index in range(len(tmp_list)):
        permutation_set.add("".join(tmp_list))

        tmp_list[index], tmp_list[ tmp_list.index(char) ] = char, tmp_list[index]

        permutation_set.add("".join(tmp_list))
    

print(
    len(permutation_set)
)
permutation_set
```

## 20. بررسی کنید رشته فقط شامل اعداد است یا خیر، بدون isnumeric().

```python
from string import digits

def is_numeric(str: str) -> bool:
    """
    Try to implement `str.isnumeric()` functionality manually.
    
    Args:
        str (str): 
    
    Returns:
        bool
    """
    for c in str:
        if not c in digits:
            return False
        else:
            continue
    
    return True

string = "1234"

print(f"string: '{string}'\n---\nimplemented is_numeric(string) => '{is_numeric(string)}'\n---\nbuilt-in isnumeric(string) => '{string.isnumeric()}'")
```

## 21. دومین عدد بزرگ یک لیست را پیدا کنید.

```python
from string import digits

sample_list = [1,2,3,4,'a','b']

def bubble_sort(iter):
    for i in range(len(iter)-1):
        for j in range(len(iter)-1-i):
            if iter[j] > iter[j+1]:
                temp = iter[j]
                iter[j] = iter[j+1]
                iter[j+1] = temp
        
    return iter

def manual_sort(list: list) -> list:
    """
    Get a list and sort it manually.

    Args:
        list (list): 
    
    Returns:
        list: sorted from big to small
    """
    
    for e in list:
        try:
            int(e)
        except ValueError:
            list[list.index(e)] = ord(e)

    return bubble_sort(list)[::-1]


print(f"list: {sample_list}\n---\nsecond biggest item in list with my way: '{manual_sort(sample_list)[1]}'\n---\nsecond biggest item in list with built-in way: '{sorted(sample_list, reverse=True)[1]}'")
```

### What is `Selection Sort` ?
> ChatGPT
>> **Selection Sort** repeatedly finds the smallest (or largest) element in the unsorted part of a list and places it in its correct position.    
>> Algorithm:    
>> 1. Find the minimum element.    
>> 2. Swap it with the first unsorted element.    
>> 3. Repeat for the remaining unsorted portion.    
>> Time Complexity: **O(n²)**, Space Complexity: **O(1)**.    


```python
from random import shuffle
sample_list = [3, 'a', 2, 1, 4, 'b']
sample_list = [1, 2, 3, 4, 'a', 'b']


def selection_sort(list: list) -> list:
    for e in list:
        try:
            int(e)
        except ValueError:
            list[list.index(e)] = ord(e)
    
    for i in range(len(list), -1, -1):
        minimum_element = list[0]
        for e in list:
            if e < minimum_element:
                minimum_element = e
            list[-i], list[list.index(minimum_element)] = minimum_element, list[-i]
        
    
    return list

selection_sort(sample_list)
```
