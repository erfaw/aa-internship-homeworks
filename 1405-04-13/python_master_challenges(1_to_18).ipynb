## 3. بدون استفاده از count() تعداد تکرار هر کاراکتر را پیدا کنید.

```python
def manual_count(_str: str) -> dict:
    """
    Get string sequence and check for count of each character in the sequence. then return a result in dict.
    Args:
        _str(str): A string sequence to check for. 
    
    Returns:
        dict: 
    """
    LEN = {}
    for c in _str:
        if not LEN.get(c):
            LEN[c] = 1
        else:
            LEN[c] += 1
    return LEN

manual_count("1223334444")
```

## 4. بررسی کنید آیا دو رشته Anagram هستند یا خیر.

### What is `Anagram`?
>>> ChatGPT :   
>>>**Anagram** means two words or phrases contain **exactly <u>the same letters</u>** with the **<u>same frequency</u>**, but arranged in a different order.    
>>> **Same frequency** means each letter <u>appears the same number of times in both words</u>.
>>> Examples: `listen` ↔ `silent` | `evil` ↔ `vile`.  
>>> Spaces, punctuation, and letter case are often ignored depending on the problem.



```python
string_1 = "evil"
string_2 = "vile"

def is_anagram(str_1: str, str_2: str) -> bool:
    """
    Check are two sequence anagram together or not. for checking use `manuall_count()` function.
    
    Args:
        str_1(str): First string sequence.
        str_2(str): Second string sequence.
    
    Returns:
        bool:
    """
    return True if manual_count(str_1) == manual_count(str_2) else False


if is_anagram(
    str_1=string_1,
    str_2=string_2,
):
    print(f" ✅ The '{string_1}' and '{string_2}' are Anagram together.")
else:
    print(f" ❌ The '{string_1}' and '{string_2}' are NOT Anagram together.")
```

## 6. طولانی‌ترین کلمه یک جمله را پیدا کنید.


```python
sentence = "I am a human with 404 on my headdddd"


def longest_word(seq: str) -> str:
    """
    Find longest word in a string sequence.

    Note:
        Maybe there is 2 or more words with longest length between words. returns the last one it find.

    Args:
        seq (str):
            A string sequence. (a sentence).

    Returns:
        str: longest word.
    """
    splited_seq = seq.split()
    splited_seq.sort(
        key= lambda x: len(x),
        reverse=True,
    )
    return splited_seq[0]


longest_word(sentence)
```

## 11. متد replace() را خودتان پیاده‌سازی کنید.

```python
sentence = 'I am a human with with with 404 on my headdddd'


def manual_replace(seq: str, sub: str, r_with: str, count: int = -1) -> str:
    """
    Try to implement functionality of `str.replace()` manually.

    Args:
        seq (str): A string sequence.
        sub (str): A string sequence to replace with.
        r_with (str): A string sequence to placed on `sub`.
        count (int): Determine how many occurance will be replaced. (default to -1, which replace all occurances)

    Returns:
        str: A string sequence replaced with args.
    """
    if not sub in seq:
        raise ValueError(f"There is not any '{sub}' in '{seq}'")
    split = seq.split()
    if count == -1:
        count = len(split)
    for _ in range(count):
        try:
            split.index(sub)
        except ValueError:
            break
        else:
            split[split.index(sub)] = r_with   

    return " ".join(split)


manual_replace(
    seq=sentence,
    sub="with",
    r_with="without",
    count=5,
)
```

## 12. متد strip() را بدون استفاده از strip() پیاده‌سازی کنید.

```python
sentence_1 = "     I am a human with    with with 404 on my headdddd         "
```

```python
# First way
def manual_strip(seq: str) -> str:
    """
    Try to implement functionality of `str.strip()` manually.

    Args:
        seq (str): A string sequence.

    Returns:
        str
    """
    return_seq = "" # iterate to a non space char. then slice from there to end. 
    for c in seq:
        if c == " ":
            continue
        else:
            return_seq = seq[seq.index(c):]
            break

    return_seq = return_seq[::-1] # reverse sequence.
    for c in return_seq: # do same as above.
        if c == " ":
            continue
        else:
            return_seq = return_seq[return_seq.index(c):]
            break

    return_seq = return_seq[::-1] # reverse again to back to normal sequence, then return.

    return return_seq


print(f"'{sentence_1.strip()}'")
print("---")
print(f"'{manual_strip(seq=sentence_1)}'")
```

```python
def better_manual_strip(seq: str) -> str:
    """
    Try to implement functionality of `str.strip()` manually.

    Note:
        This one using pop() method of lists. also, works with inner function.

    Args:
        seq (str): A string sequence.

    Returns:
        str
    """
    def iter_to_char(seq: str) -> str:
        rseq = list(seq)
        for i in range(len(seq)-1, -1, -1):
            if seq[i] == " ":
                rseq.pop()
            else:
                break
        return "".join(rseq[::-1])

    return iter_to_char(iter_to_char(seq))


print(f"'{sentence_1.strip()}'")
print("---")
print(f"'{manual_strip(seq=sentence_1)}'")
```

## 15. بررسی کنید رشته Palindrome است یا خیر.

### What is `Palindrome`
>>> ChatGPT  
>>> Palindrome is a word, phrase, or sequence that <u>reads the same forward and backward</u>. Examples: madam - racecar - 1221

```python
def is_palindrome(word: str) -> bool:
    """
    Check a word for palindrome.

    Note:
        Palindrome is a word, phrase, or sequence that reads the same forward and backward.

    Args:
        word (str): a word as a string sequence.

    Returns:
        bool
    """
    d = (len(word) / 2).__ceil__()
    first_half = word[: d + 1]
    word = word[::-1]
    second_half = word[: d + 1]
    return first_half == second_half


is_palindrome(
    word="racecar",
)
```

## 17. بزرگ‌ترین زیررشته بدون کاراکتر تکراری را پیدا کنید.

```python
from string import ascii_letters

sentence = "i will give him an offer he can't refuse."

sub_sequence = set()

for i in range(len(sentence)):
    for j in range(len(sentence)):
        sub_sequence.add(sentence[j:i+1])
sub_sequence.remove('')

sub_sequence_list = list(sub_sequence)
sub_sequence_list.sort(key= lambda x: len(x), reverse=True)

longest_non_repeated_sub_sequence = None
for sub_seq in sub_sequence_list:
    is_repeated = False
    for c in ascii_letters:
        if sub_seq.count(c) > 1:
            is_repeated = True
            break
    if is_repeated:
        continue
    else:
        longest_non_repeated_sub_sequence = sub_seq
        break

longest_non_repeated_sub_sequence
```
