## 1. برنامه‌ای بنویسید که بدون استفاده از `str[::-1]`یک رشته را برعکس کند.

```python
str_example = "Hello World from Pythonistas! <3 1"
```

```python
# First way
for char in reversed(str_example):
    print(char, end='')
```

```python
# Second way
reversed_str = list(str_example)
r_i = len(str_example)-1
for i in range(len(str_example)):
    if not r_i <= i:
        reversed_str[i], reversed_str[r_i] = reversed_str[r_i], reversed_str[i]
        r_i -= 1

print(
    "".join(reversed_str)
)
```

## 2. تعداد حروف بزرگ، حروف کوچک، اعداد و کاراکترهای خاص یک رشته را محاسبه کنید.

```python
from string import punctuation, ascii_uppercase, ascii_lowercase

str_example = "1234abcdABCD,!#@" # a string contains four char for each type.

LEN = {
    "_number": 0,
    "_lower_string": 0,
    "_upper_string": 0,
    "_special_char": 0,
}

for c in str_example:
    if c.isdecimal():
        LEN["_number"] += 1
        continue
    elif c in ascii_uppercase:
        LEN["_upper_string"] += 1 
        continue
    elif c in ascii_lowercase:
        LEN["_lower_string"] += 1
        continue
    elif c in punctuation:
        LEN["_special_char"] += 1 
    else:
        print('There is a Problem!')

LEN
```

## 3. بدون استفاده از count() تعداد تکرار هر کاراکتر را پیدا کنید.

```python
def manual_count(_str: str) -> dict:
    """
    Get string sequence and check for count of each character in the sequence. then return a result in dict.
    Args:
        _str(str): A string sequence to check for. 
    
    Returns:
        dict: 
    """
    LEN = {}
    for c in _str:
        if not LEN.get(c):
            LEN[c] = 1
        else:
            LEN[c] += 1
    return LEN

manual_count("1223334444")
```

## 4. بررسی کنید آیا دو رشته Anagram هستند یا خیر.

### What is `Anagram`?
>>> ChatGPT :   
>>>**Anagram** means two words or phrases contain **exactly <u>the same letters</u>** with the **<u>same frequency</u>**, but arranged in a different order.    
>>> **Same frequency** means each letter <u>appears the same number of times in both words</u>.
>>> Examples: `listen` ↔ `silent` | `evil` ↔ `vile`.  
>>> Spaces, punctuation, and letter case are often ignored depending on the problem.



```python
string_1 = "evil"
string_2 = "vile"

def is_anagram(str_1: str, str_2: str) -> bool:
    """
    Check are two sequence anagram together or not. for checking use `manuall_count()` function.
    
    Args:
        str_1(str): First string sequence.
        str_2(str): Second string sequence.
    
    Returns:
        bool:
    """
    return True if manual_count(str_1) == manual_count(str_2) else False


if is_anagram(
    str_1=string_1,
    str_2=string_2,
):
    print(f" ✅ The '{string_1}' and '{string_2}' are Anagram together.")
else:
    print(f" ❌ The '{string_1}' and '{string_2}' are NOT Anagram together.")
```

## 6. طولانی‌ترین کلمه یک جمله را پیدا کنید.


```python
sentence = "I am a human with 404 on my headdddd"


def longest_word(seq: str) -> str:
    """
    Find longest word in a string sequence.

    Note:
        Maybe there is 2 or more words with longest length between words. returns the last one it find.

    Args:
        seq (str):
            A string sequence. (a sentence).

    Returns:
        str: longest word.
    """
    splited_seq = seq.split()
    splited_seq.sort(
        key= lambda x: len(x),
        reverse=True,
    )
    return splited_seq[0]


longest_word(sentence)
```

## 7. اولین کاراکتری که فقط یک بار تکرار شده را پیدا کنید.

```python
sentence = "I am a human with 404 on my headdddd"

just_1_repeat = []
for k, v in manual_count(sentence).items():
    if v == 1:
        just_1_repeat.append(k)

first_c = len(sentence)+1
for c in just_1_repeat:
    if first_c > sentence.index(c):
        first_c = sentence.index(c)

print(sentence + "\nand the first one is:")
sentence[first_c]
```

## 8. اولین کاراکتری که بیش از یک بار تکرار شده را پیدا کنید.

## 9. بدون استفاده از split() رشته را به کلمات تقسیم کنید.

```python
sentence = "I am a human with 404 on my headdddd"
print(f"this is buil-in split() method result:\n\t{sentence.split()}")

def manual_split(seq: str, by: str=" ") -> list[str]: 
    """
Try to implement functionality of buil-in `str.split()` in python. 

    Args: 
        seq (str): a string sequence.
        by (str): a character to split by. (default to ' ' empty space.)

    Returns: 
        list[str]: list of splited by strings.
    """
    splited_without_split_func = []
    word=""
    for c in sentence:
        if not c == by:
            word += c
        else: 
            splited_without_split_func.append(word)
            word=""
    splited_without_split_func.append(word)

    return splited_without_split_func


print(f"this is manual implementation of split() by me:\n\t{manual_split(sentence)}")
```
