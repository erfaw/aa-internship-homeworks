## 1. برنامه‌ای بنویسید که بدون استفاده از `str[::-1]`یک رشته را برعکس کند.

```python
str_example = "Hello World from Pythonistas! <3 1"
```

```python
# First way
for char in reversed(str_example):
    print(char, end='')
```

```python
# Second way
reversed_str = list(str_example)
r_i = len(str_example)-1
for i in range(len(str_example)):
    if not r_i <= i:
        reversed_str[i], reversed_str[r_i] = reversed_str[r_i], reversed_str[i]
        r_i -= 1

print(
    "".join(reversed_str)
)
```
